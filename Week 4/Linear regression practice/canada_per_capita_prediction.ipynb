```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.scatter(df['year'], df['per capita income (US$)'], color='red')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Over Time')
plt.show()
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(df[['year']],df['per capita income (US$)'])
```

```python
model.predict(pd.DataFrame([[2025]], columns=['year']))
```

```python
sns.regplot(x=df['year'],y=df['per capita income (US$)'],data=df)
```
